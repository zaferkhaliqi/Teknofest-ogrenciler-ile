```python
import cv2
import numpy as  np
import matplotlib.pyplot as plt
from cvzone.HandTrackingModule import HandDetector
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
detector = HandDetector(maxHands=2)
cap = cv2.VideoCapture(0)

def treshold (imgCrop):

    gry = cv2.cvtColor(imgCrop,cv2.COLOR_BGR2GRAY)
    ret,img = cv2.threshold(gry,200,255,cv2.THRESH_BINARY)
    return img

def ikiEliKirp (hands,img):
    offset = 20
    bbox_right = list(hands[0]['bbox'])# Sağ elin sınırlayıcı kutusunu içeren bir liste oluşturuluyor.
    bbox_left = list(hands[1]['bbox'])# Sol elin sınırlayıcı kutusunu içeren bir liste oluşturuluyor.
    min_x = min(bbox_right[0], bbox_left[0])
    max_x = max(bbox_right[0] + bbox_right[2], bbox_left[0] + bbox_left[2])# sağ ve sol ellerin sınırlayıcı kutularının maximum x ve y koordinatları belirleniyor

    min_y = min(bbox_right[1], bbox_left[1])
    max_y = max(bbox_right[1] + bbox_right[3], bbox_left[1] + bbox_left[3])
    imgCrop = img[min_y-offset:max_y+offset, min_x-offset:max_x+offset]# maximum x ve y,minimum x ve y değerlerine belirlediğimiz ofset değerini maximuma toplama minimuma çıkarma işlemi yapılıyor
    return imgCrop

def tekElKirpma (hands,img):
    x,y,w,h = hands[0]['bbox']
    offset = 20
    yedekgoruntu =  img[y-offset: y+h+offset , x-offset:x+w +offset]
    return yedekgoruntu

def sinirAgI ():


    num_classes = 29

    siniragi = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(400, 400, 1)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    return siniragi
class_labels = [
    'A', 'B', 'C', 'Ç', 'Ş', 'D', 'E', 'F',
    'G', 'H', 'İ', 'J', 'K', 'L', 'M', 'N', 'O',
    'Ü', 'I', 'O', 'P', 'R', 'S', 'T', 'U',
    'V', 'Y', 'Ğ', 'Z'
]
def test(tresholdimg, siniragi):
    # Preprocess the image
    img = tresholdimg
    # img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert the image to grayscale

    # Normalize the image data
    img_array = np.expand_dims(img, axis=0)  # Add batch dimension

    # Use the model to make a prediction
    prediction = siniragi.predict(img_array)
    predicted_class_index = np.argmax(prediction)

    # Define class labels (you can customize these based on your problem)


# Get the predicted class label
    predicted_class_label = class_labels[predicted_class_index]

    print(f"Predicted class: {predicted_class_label}")
    return predicted_class_label

# Load the model
siniragi_model = sinirAgI()
siniragi_model.load_weights(r'C:\Users\zafer\OneDrive\Desktop\siniragi_agirliklari_04.h5')

# Load the image using OpenCV
```

```python
detector = HandDetector(maxHands=2)
cap = cv2.VideoCapture(0)

while True:
    try:
        result,img = cap.read()
        hands,img = detector.findHands(img)
        if hands:
            if len(hands)==1:
                imgCrop = tekElKirpma(hands=hands,img = img)
                tresholdedImg = treshold(imgCrop = imgCrop)

            elif len(hands)==2:
                imgCrop = ikiEliKirp(hands=hands,img=img)
                tresholdedImg = treshold(imgCrop = imgCrop)
            resized = cv2.resize(tresholdedImg,(400,400))
            resized = resized/255.0
            toplot = cv2.flip(img,1)

            cv2.imshow('Genel',toplot)
            cv2.imshow('treshed',resized)
            key = cv2.waitKey(1)
            test_sonucu = test(tresholdimg=resized, siniragi=siniragi_model)

            print(test_sonucu)
            
            if key == ord('q'):
                cv2.destroyAllWindows()
                break
            
    except:
        print()
```
